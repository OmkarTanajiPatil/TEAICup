## Seperating variable_attribute Feature:
* lowerLimit: -0.06499, upperLimit: 0.00499, nominalValue: 0.53, engineeringUnits: mm, cavity: 1

```python
# We can use regex to map and seperate -> Remember for later use
```

```python
def seperate_variable_attribute_feature(X:str):
    variable_attribute_list = X.split(",")
    variable_attribute_map = {}
    for i in variable_attribute_list:
        name , value = i.split(':')
        variable_attribute_map[name.strip()] = value.strip()
    return pd.Series(variable_attribute_map)
```

```python
d2 = d2.join(d2.variable_attribute.apply(seperate_variable_attribute_feature))
d2.head()
```

```python
# Now making these features float -> (Before they are object)
cols = ["lowerLimit", "upperLimit","nominalValue", "cavity"]
d2[cols] = d2[cols].astype(float)
d2.dtypes
```
