## Matching the variable names to match other datasets


```python
def matchVariables(x):
    x = x.split(" ")
    x = [i for i in x if i not in ("", " ")]
    x = [i.split("-") for i in x]
    x = [item for sublist in x for item in sublist]
    x = [i.capitalize() for i in x]

    x[0] = x[0].lower()

    x = "".join(x)
    return x
```

```python
matchVariables(" Teil 2 Gap Y ")
```

```python
d3.error_variable = d3.error_variable.apply(matchVariables)
d3
```
