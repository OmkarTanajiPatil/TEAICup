## Some Basic Filters
+ Removing Single valued attribute
+ Machine code 200

```python
# Removing single valued attribute

single_valued_columns_list = ["hydra_plant_instance_name"]

for i in single_valued_columns_list:
    if i in d1.columns:
        d1.drop(i, inplace= True, axis = 1)
```

```python
d1
```

```python
# Filtering status code 200

d1 = d1[d1["machine_status_code"] == 200]
d1
```

```python
# Now machine status code not important removing it also
single_valued_columns_list = ["machine_status_code", "machine_status_name"]

for i in single_valued_columns_list:
    if i in d1.columns:
        d1.drop(i, inplace=True, axis=1)
d1
```

```python
# Converting the time to timestamp and sorting based on start time
time_columns = ["machine_event_start_timestamp", "machine_event_start_end_timestamp"]
for i in time_columns:
    if i in d1.columns:
        d1[i] = pd.to_datetime(d1[i])
d1
```

```python
d1.sort_values(by="machine_event_start_timestamp", inplace=True)
d1
```

```python
# Saving this filtered data in parquet
d1.to_parquet(r"E:\Learning\TEAI Cup\Pandas\Preprocessing D1\d1_1_general_filtering.parquet", index=False)
```
