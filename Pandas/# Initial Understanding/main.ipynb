### 3. Combining the Data
* Current Process
    * Data is filtered based on Machine Code: 200
    * Data is filtered based on machine id
    * Data is filtered based on timestamp

```python
def getData(machine_id, start_time, end_time):
    base_path = "newCombinedDatasets"
    
    machine_id = machine_id.split("-")[-1]
    
    formatted_start_time = start_time.strftime("%Y-%m-%d_%H-%M-%S")
    formatted_end_time   = end_time.strftime("%Y-%m-%d_%H-%M-%S")

    
    new_folder = f"Id_{machine_id}_start_{formatted_start_time}_end_{formatted_end_time}"
    
    new_path = os.path.join(base_path, new_folder)
    os.makedirs(new_path, exist_ok=True)
    
    camera_data = d2[
        (d2.device_name == f"de512stmp{machine_id}")
        & (d2.timestamp.between(start_time, end_time))
    ]
    camera_data.sort_values(by="timestamp", inplace= True)
    
    error_data = d3[
        (d3.qcpc_event_machine_id == f"S-{machine_id}")
        & (d3.qcpc_event_create_timestamp.between(start_time, end_time))
    ]
    
    error_data.sort_values(by="qcpc_event_create_timestamp", inplace=True)
    
    camera_data.to_csv(
        os.path.join(
            new_path,
            f"Camera_data_Id_{machine_id}_start_{formatted_start_time}_end_{formatted_end_time}.csv",
        ),
        index=False
    )
    print("Camera data saved Successfully!")
    
    error_data.to_csv(
        os.path.join(
            new_path,
            f"Error_data_Id_{machine_id}_start_{formatted_start_time}_end_{formatted_end_time}.csv"
        )
        , index=False
    )
    print("Error data saved Successfully!")
    return camera_data, error_data
```

```python
sample = d1[d1.machine_status_code == 200].sample()
start_time = sample.machine_event_start_timestamp.iloc[0]
end_time = sample.machine_event_start_end_timestamp.iloc[0]
machine_id = sample.machine_id.iloc[0]
print(
    f'''Machine id : {sample.machine_id.iloc[0]}\nstart = '{sample.machine_event_start_timestamp.iloc[0]}'\nend = '{sample.machine_event_start_end_timestamp.iloc[0]}' \nTotal Time Spend : {end_time - start_time}'''
)
```

```python
# Using the function to get the sorted data
camera_data, error_data = getData(machine_id, start_time, end_time)
```
