```python
import pandas as pd


def split_measurement(text):
    out = pd.Series(text).str.extract(
        r"^(.*?):\s*Cur=\s*(.*?)\s+LTL=\s*(-?[0-9.]+)\s+UTL=\s*(-?[0-9.]+)"
    )
    out.columns = ["error_in", "cur", "ltl", "utl"]
    row = out.iloc[0]

    # numeric conversion (safe)
    row["ltl"] = float(row["ltl"])
    row["utl"] = float(row["utl"])
    row["cur"] = pd.to_numeric(row["cur"], errors="coerce")

    return row


# usage
df = pd.DataFrame(
    {
        "raw": [
            "Teil 1 Gap Y: Cur= 0.469 LTL= 0.470 UTL= 0.530",
            "SP 1 durchgebrannt GW 70: Cur= 45.405 LTL= 0.000 UTL= 45.000",
            "Spur 2 Gap: Cur= Nicht messbar LTL= 0.160 UTL= 0.220",
            " SP1 DL Erkennung Ergebnis: Cur= 67572.956 LTL= -0.100 UTL= 65000.000"
        ]
    }
)

df[["error_in", "cur", "ltl", "utl"]] = df["raw"].apply(split_measurement)
print(df)
```
