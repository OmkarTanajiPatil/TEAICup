## Some Basic Filters
+ Removing Single valued attribute
+ Machine code 200

```python
single_valued_columns = ["hydra_plant_instance_name"]

for col in single_valued_columns:
    if col in df.columns:
        df = df.drop(col)
df
```

```python
df = df.filter(pl.col("machine_status_code") == 200)
df
```

```python
single_valued_columns = ["machine_status_name", "machine_status_code"]

for col in single_valued_columns:
    if col in df.columns:
        df = df.drop(col)
df
```

### Converting times into timestamp

```python
cols_to_timestamp = [
    "machine_event_start_timestamp",
    "machine_event_start_end_timestamp",
]
df = df.with_columns(
    pl.col(cols_to_timestamp).str.to_datetime()
)
df
```

### Sorting based on start time

```python
df = df.sort("machine_event_start_timestamp")
df
```
