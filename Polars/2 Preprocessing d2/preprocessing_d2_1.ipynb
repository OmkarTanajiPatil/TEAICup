## Removing unnecessory data features

```python
unnecessory_features = ["device", "machine_def", "year", "month", "date"]
for feature in unnecessory_features:
    if feature in df.columns:
        df = df.drop(pl.col(feature))
df.head()
```

# Converting to timestamp

```python
if df["timestamp"].dtype == pl.String:
    df = df.with_columns(
        pl.col("timestamp").str.to_datetime()
    )
df.head()
```

## Seperating variable_attribute Feature:
* lowerLimit: -0.06499, upperLimit: 0.00499, nominalValue: 0.53, engineeringUnits: mm, cavity: 1

```python
keys = ["lowerLimit", "upperLimit", "nominalValue", "engineeringUnits", "cavity"]

df = df.with_columns(
    pl.col("variable_attribute")
    .str.split(",")
    .list.eval(
        pl.element()
        .str.split(":")
        .list.eval(
            pl.element().str.strip_chars()
        )
    )
    .list.eval(
        pl.element()
        .list.last()
    )
    .list.to_struct(fields=keys)

).unnest("variable_attribute")
df
```

## Removing unnecessory data features

```python
unnecessory_features = ["engineeringUnits", "cavity", "machine_definition"]
for feature in unnecessory_features:
    if feature in df.columns:
        df = df.drop(pl.col(feature))
df.head()
```

## Making new features float

```python
df = df.with_columns(
    pl.col(['lowerLimit', 'upperLimit', 'nominalValue']).cast(pl.Float64)
)
df
```

## Converting the device name to name matching other datasets


```python
df = df.with_columns((
    "S-" + 
    pl.col("device_name")
    .str.slice(-3))
    .alias("device_name")
)
df
```

### Renaming the device_name : machine_id

```python
df = df.rename({'device_name': 'machine_id'})
df
```

```python
df = df.with_columns(
    pl.col("value").str.extract_all(r"-?\d+\.?\d*").cast(pl.List(pl.Float64))
)
df
```

# Adding some feature to dataset to detect the variations in value field

```python
lf = df.lazy()
lf
```

```python
vals = (
    pl.col('value')
    .str.split(",")
    .list.eval(
        pl.element().cast(
            pl.Float64
        )
    )
)
```

```python
outside_range_checks = [
    (vals.list.get(i) > pl.col("upperLimit"))
    | (vals.list.get(i) < pl.col("lowerLimit"))
    for i in range(100)
]
```

```python
lf = lf.with_columns(
    [
        pl.col("value")
        .str.extract_all(r"-?\d+\.?\d*")
        .cast(pl.List(pl.Float64))
        .list.min()
        .alias("min")
    ]
)
```

```python
df = lf.collect(streaming=True)
df
```
